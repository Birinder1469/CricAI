```python
#Import CSV File
import pandas as pd
from sklearn.model_selection import train_test_split as tts
match_data = pd.read_csv('finalDataset.csv')
match_data.head()
```

```python
#Display Columns
for i, feature in enumerate(list(match_data.columns)):
    print (i, feature)
```

```python
#Separation into Feature Set & Target Set

target_cols = match_data.columns[212:232]
feature_cols = match_data.columns[0:212]

print(target_cols)
print(feature_cols)

target_set = match_data[target_cols]
feature_set = match_data[feature_cols]

print(len(target_set))
print(len(feature_set))
```

```python
#Splitting Into Training & Testing Datasets
X = feature_set
Y = target_set
X_train, X_test, Y_train, Y_test = tts(X, Y, test_size=0.25)
print(len(X_train))
print(len(Y_train))
```

```python
#Decision Tree Classifier 1

from sklearn import tree
DTclf1 = tree.DecisionTreeClassifier()
DTclf1.fit(X_train, Y_train)
```

```python
predicted = DTclf1.predict(X_test)
predicted
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, predicted))
```

```python
train_prediction = DTclf1.predict(X_train)
accuracy_score(Y_train, train_prediction)
```

```python
#Hash Indexing Teams 1

ourTeams_1 = {}
for i,teams in enumerate(list(match_data.columns)):
    if i==23:
        break
    teamName = str(teams)
    teamName = teamName.split("Team 1_", 1)[-1]
    ourTeams_1[teamName] = i
print(ourTeams_1)
```

```python
#Hash Indexing Teams 2

ourTeams_2 = {}
for i,teams in enumerate(list(match_data.columns)):
    if i<23:
        continue
    if i==46:
        break
    teamName = str(teams)
    teamName = teamName.split("Team 2_", 1)[-1]
    ourTeams_2[teamName] = i
print(ourTeams_2)
```

```python
#Hash Indexing Grounds & Innings

ourGrounds = {}
for i,grounds in enumerate(list(match_data.columns)):
    if i<46:
        continue
    if i==207:
        break
    groundName = str(grounds)
    groundName = groundName.split("Ground_", 1)[-1]
    ourGrounds[groundName] = i
print(ourGrounds)
print("\n")
ourInnings = {'1': 207, '2': 208}
print(ourInnings)
print("\n")
ourVenues = {'Home': 209, 'Away': 210, 'Neutral': 211}
print(ourVenues)
```

```python
#Hash Indexing the Target Winners

winnerIndex = {}
counter = 0
for i,teams in enumerate(list(match_data.columns)):
    if i<212:
        continue
    
    teamName = str(teams)
    teamName = teamName.split("Winner_", 1)[-1]
    winnerIndex[teamName] = counter
    counter+=1
print(winnerIndex)
```

```python
#Running Prediction in Detailed Manner

inputPrediction = [0]*212
#print(inputPrediction)
t1 = 'Pakistan'
t2 = 'India'
ground = 'Mumbai'
innings = '1'
venue = 'Neutral'
inputPrediction[ourTeams_1[t1]] = 1
inputPrediction[ourTeams_2[t2]] = 1
inputPrediction[ourGrounds[ground]] = 1
inputPrediction[ourInnings[innings]] = 1
inputPrediction[ourVenues[venue]] = 1
#print(inputPrediction)
print("\n\n\n")

#Running Predictions

ourPrediction = DTclf1.predict([inputPrediction])
print(ourPrediction)
#ourProb = DTclf1.proba([inputPrediction])
#print(ourProb)
```
